## Data Split

```python
x = data.drop('SalePrice',axis=1)
y = data['SalePrice']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=41,)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,BaggingRegressor,HistGradientBoostingRegressor
```

```python
models = {'LinearR':LinearRegression(),
          'Ridge':Ridge(),
          'Lasso':Lasso(),
          'Tree':DecisionTreeRegressor(),
          'KNN':KNeighborsRegressor(),
          'RandomF':RandomForestRegressor(),
          'GBR':GradientBoostingRegressor(),
          'Bagg':BaggingRegressor(),
          'HGBR':HistGradientBoostingRegressor()}
```

```python
def fit_score(models):
  '''
  This Funcytion will show the score on training 
  '''

  np.random.seed(42)

  train_score = {}
  test_score = {}

  for name,model in models.items():
    model.fit(x_train,y_train)

    train_score[name] = model.score(x_train,y_train)

    test_score[name] = model.score(x_test,y_test)

  return train_score,test_score
    


```

```python
train_score,test_score = fit_score(models)
```

```python
train_score
```

```python
test_score
```

```python
train_scores = pd.DataFrame(train_score,index =['training'])
test_scores = pd.DataFrame(test_score,index=['test'])
```

```python
scores = pd.concat([train_scores,test_scores],axis=0)
```

```python
scores
```

```python
scores.T.plot.bar(figsize=(15,6));
```

as we can see Linear Regression , Random Forest & Gradient Boosting Regressor models are performing well, so we'll remove other models &  check these models with new metrics to find which one is better fit

```python
models = {'LinearR':LinearRegression(),
          'RandomF':RandomForestRegressor(),
          'GBR':GradientBoostingRegressor()}
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def show_score(model):
  """
  This Function will show the scores in Training & Testing Data
  """

  train_preds = model.predict(x_train)
  test_preds = model.predict(x_test)

  scores = {'Training MAE':mean_absolute_error(y_train,train_preds),
            'Test MAE':mean_absolute_error(y_test,test_preds),
            'Training RMSE':np.sqrt(mean_squared_error(y_train,train_preds)),
            'Test RMSE':np.sqrt(mean_squared_error(y_test,test_preds)),
            'Training R^2 Score':r2_score(y_train,train_preds),
            'Test R^2 Score':r2_score(y_test,test_preds)}

  return scores
```

```python
score_dict = {}
for name,model in models.items():
  model.fit(x_train,y_train)

  score_dict[name] = show_score(model)
```

```python
def plot_features(columns,importance,n):
  df = pd.DataFrame({"feature":columns,'feature_importance':importance}).sort_values('feature_importance',ascending=True)

  # Plot the DataFrame
  plt.figure(figsize=(7,10))
  plt.barh(df['feature'][:n],df['feature_importance'][:n])
  plt.xlabel('Feature Importance')
  plt.ylabel('Feature')
  plt.title('Feature Imporatnce Visual')
```
